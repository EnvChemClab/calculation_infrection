```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import os

df = pd.read_csv("./temp2lnresist.csv", encoding="UTF-8")
plt.figure()
plt.grid()

temp = df["1/T"]
resist = df["lnR"]

plt.scatter(temp, resist, c="red")

func =  np.poly1d(np.polyfit(temp, resist, 1))

plt.xlabel("1/(Temperature)[1/℃]")
plt.ylabel("ln(Resistance)[lnΩ]")
plt.plot(temp,func(temp), "b", marker=".")

print(func)
```

```python
%matplotlib inline
import re
import numpy as np
from sympy import * 
from scipy import optimize as opt
import matplotlib.pyplot as plt
import pandas as pd
import os

df = pd.read_csv("./resist2temp.csv", encoding="UTF-8")
plt.figure()
plt.grid()

temp = df["T"]
resist = df["R"]

plt.scatter(temp, resist, c="red")

func = np.poly1d(np.polyfit(temp, resist, 3))
plt.plot(temp, func(temp), "b", marker=".")

value = 0
beforeSub = 1

while value < 500:
    x = symbols("x")
    solution = diff(func(x), x, 2)
    substitutioned = solution.subs(x, value)
    #print(substitutioned)
    if beforeSub > abs(substitutioned):
        beforeSub = abs(substitutioned)
        infrection = value
    
    value += 0.1
   
print(func)
print(solution)
print(beforeSub)
print(infrection)
```
